```python
import pandas as pd
df = pd.read_csv("investimento.csv")
df
```

```python
df.plot(kind="scatter", x="tempo", y="renda");
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["tempo"]]
y = df[["renda"]]

modelo.fit(x, y)
```

```python
modelo.predict([[101]])[0][0]
```
